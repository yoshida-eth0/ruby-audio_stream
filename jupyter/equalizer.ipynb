```ruby
$LOAD_PATH << File.dirname(__FILE__) + "/../lib"
require 'audio_stream'

include AudioStream
include AudioStream::Fx

soundinfo = SoundInfo.new(
  channels: 2,
  samplerate: 44100,
  window_size: 1024,
  format: RubyAudio::FORMAT_WAV|RubyAudio::FORMAT_PCM_16
)
```

```ruby
eq = Equalizer2band.new(soundinfo, lowgain: 3, highgain: 5)
eq.plot.show
nil
```

```ruby
eq = Equalizer3band.new(soundinfo, lowgain: 3, midgain: 1, highgain: 5)
eq.plot.show
nil
```

```ruby
eq = GraphicEqualizer.new(soundinfo)
  .add(freq: 100, gain: 8)
  .add(freq: 200, gain: 4)
  .add(freq: 400, gain: -4)
  .add(freq: 800, gain: -8)
  .add(freq: 1600, gain: 4)
  .add(freq: 3200, gain: 8)
eq.plot.show
nil
```
