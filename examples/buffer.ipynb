```ruby
$LOAD_PATH << File.dirname(__FILE__) + "/../lib"
require 'audio_stream/core_ext'

include AudioStream
include AudioStream::Fx

soundinfo = SoundInfo.new(
  channels: 2,
  samplerate: 44100,
  window_size: 1024,
  format: RubyAudio::FORMAT_WAV|RubyAudio::FORMAT_PCM_16
)

sin_wave = AudioInput.sin(440.0, 100, soundinfo: soundinfo).each.next
sin_wave.plot(soundinfo).wave.show
```
