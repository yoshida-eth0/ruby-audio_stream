```ruby
$LOAD_PATH << File.dirname(__FILE__) + "/../lib"
require 'audio_stream/core_ext'

include AudioStream::Synth

xs = []
ys1 = []
ys2 = []
ys3 = []

pos = ShapePos.new
step = 0.01

(1/step).to_i.times {|i|
  phase = pos.next(step)
  xs << i * step
  ys1 << Modulation::Curve::Straight[phase]
  ys2 << Modulation::Curve::EaseIn[phase]
  ys3 << Modulation::Curve::EaseOut[phase]
}

traces = [
  {x: xs, y: ys1, name: "Straight"},
  {x: xs, y: ys2, name: "EaseIn"},
  {x: xs, y: ys3, name: "EaseOut"},
]
Plotly::Plot.new(data: traces).show
```
