```ruby
$LOAD_PATH << File.dirname(__FILE__) + "/../lib"
require 'audio_stream'

include AudioStream
include AudioStream::Fx

soundinfo = SoundInfo.new(
  channels: 2,
  samplerate: 44100,
  window_size: 1024,
  format: RubyAudio::FORMAT_WAV|RubyAudio::FORMAT_PCM_16
)
```

```ruby
bands = 16.times.map {|i|
  6.875 * (2 ** (i*0.5+3))
}   
bandwidth = 0.25

data = []

bands.each {|band|
  bpf = BandPassFilter.create(soundinfo, freq: band, bandwidth: bandwidth)
  pd = bpf.plot_data
  data << {x: pd[:x], y: pd[:magnitude]}
}

Plotly::Plot.new(
  data: data,
  layout: {xaxis: {type: 'log'}},
).show
nil
```
